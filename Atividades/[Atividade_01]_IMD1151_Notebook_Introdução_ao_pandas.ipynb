### Seleção de dados com DataFrame

Existem dois métodos no pandas que permite selecionar parte do DataFrame: [DataFrame.loc](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.loc.html) e [DataFrame.iloc](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iloc.html).


*   O método **df.loc[]** permite acesso aos dados através dos rótulos (labels) das linhas e colunas.
*   O método **df.iloc[]** permite acesso aos dados através dos índices das linhas e colunas.


Exemplos:

```
df.loc['row1', 'col4']
```

ou

```
df.iloc[0,3]
```

````python
print(f500.iloc[0,2])
````

````python
f500.iloc[:,2:6].head()
````

### Atribuição de valores no Dataframe:

Quando usamos NumPy, as mesmas operações de seleção podem ser usadas para atribuição. Veja o exemplo abaixo:

```python
my_array = np.array([1, 2, 3, 4])

# seleção
print(my_array[0])

# atribuição
my_array[0] = 99
```

O mesmo ocorre no **pandas**. Veja o exemplo abaixo:

```python
>>> top5_rank_revenue = f500[["rank", "revenues"]].head()

>>> print(top5_rank_revenue)
                              rank  revenues
    Walmart                      1    485873
    State Grid                   2    315199
    Sinopec Group                3    267518
    China National Petroleum     4    262573
    Toyota Motor                 5    254694

>>> top5_rank_revenue["revenues"] = 0

>>> print(top5_rank_revenue)
                              rank  revenues
    Walmart                      1         0
    State Grid                   2         0
    Sinopec Group                3         0
    China National Petroleum     4         0
    Toyota Motor                 5         0
    
```
Quando selecionamos uma coluna inteira por rótulo (label) e realizamos uma atribuição, o valor é atribuído para cada item da coluna.

É possível atribuir para linha e coluna específica.

```python
>>> top5_rank_revenue.loc["Sinopec Group", "revenues"] = 999

>>> print(top5_rank_revenue)
                              rank  revenues
    Walmart                      1         0
    State Grid                   2         0
    Sinopec Group                3       999
    China National Petroleum     4         0
    Toyota Motor                 5         0
```

Se atribuirmos um valor usando um índice ou rótulo de coluna **que não existe**, o pandas criará uma nova linha ou coluna em nosso dataframe. Veja o exemplo abaixo:

```python
>>> top5_rank_revenue["year_founded"] = 0

>>> print(top5_rank_revenue)

                              rank  revenues  year_founded
    Walmart                      1         0             0
    State Grid                   2         0             0
    Sinopec Group                3       999             0
    China National Petroleum     4         0             0
    Toyota Motor                 5         0             0

>>> top5_rank_revenue.loc["My New Company"] = 555

>>> print(top5_rank_revenue)

                              rank  revenues  year_founded
    Walmart                      1         0             0
    State Grid                   2         0             0
    Sinopec Group                3       999             0
    China National Petroleum     4         0             0
    Toyota Motor                 5         0             0
    My New Company             555       555           555
```

### Explorar o Dataframe

Uma forma de explorar os objetos Series ou Dataframe do **pandas** é usando o método **.describe()**. Esse método retorna algumas estatísticas descritivas sobre os dados no objeto, seja [Series.describe()](http://pandas.pydata.org/docs/reference/api/pandas.Series.describe.html) ou [DataFrame.describe()](http://pandas.pydata.org/docs/reference/api/pandas.DataFrame.describe.html).

O exemplo abaixo mostra as informações sobre os dados de '**revenues**' do dataframe. O método deve retornar a quantidade de valores não nulos contidos no objeto, a média e o desvio padrão, juntamente com os valores mínimo, máximo e quartil.

````python
revs = f500["revenues"]
revs.describe()
````

O método **.describe()** também é aplicado para valores não numéricos. Veja o exemplo abaixo que apresenta as estatísticas descritivas da coluna **'country'**.

````python
f500["country"].describe()
````

Além da quantidade de valores não nulos, o método retornou 3 valores sobre os dados:

* **unique**: quantidade de valores diferentes no conjunto de dados.
* **top**: o valor mais frequente no conjunto de dados.
* **freq**: a quantidade de amostras do valor mais frequente. Nesse caso, o valor "USA".

---

O método .**describe()** pode ser aplicada a todo o conjunto de dados.

```
df.describe()
```
